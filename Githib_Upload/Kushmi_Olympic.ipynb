1. Dictionary Dataset Preprocessing


```python
# Load the Dictionary dataset
dictionary_df = pd.read_csv(dictionary_path)

# Step 1: Handle Missing Values
# - Fill missing values in "Population" and "GDP per Capita" columns with median values for simplicity.
dictionary_df['Population'] = dictionary_df['Population'].fillna(dictionary_df['Population'].median())
dictionary_df['GDP per Capita'] = dictionary_df['GDP per Capita'].fillna(dictionary_df['GDP per Capita'].median())

# Step 2: Standardize Country Codes and Country Names
# - Remove any trailing spaces in "Code" and "Country" columns for consistency.
dictionary_df['Code'] = dictionary_df['Code'].str.strip()
dictionary_df['Country'] = dictionary_df['Country'].str.strip()

# Step 3: Remove Duplicates (if any)
dictionary_df.drop_duplicates(inplace=True)

# Check the processed data
print("Processed Dictionary Data:\n", dictionary_df.head())
```

2. Summer Dataset Preprocessing


```python
# Load the Summer dataset
summer_df = pd.read_csv(r"C:\Users\kushi\Olympic\Olympic_Last_attemp\archive\summer.csv")

# Step 1: Handle Missing Values in "Country" Column
# - Since only four rows are missing "Country" information, we can drop these rows for simplicity.
summer_df = summer_df.dropna(subset=['Country'])

# Step 2: Standardize Country Codes
# - Strip any spaces from the "Country" column.
summer_df['Country'] = summer_df['Country'].str.strip()

# Step 3: Remove Duplicates
# - Drop duplicate rows if any (e.g., sometimes athlete names may be duplicated due to event details).
summer_df.drop_duplicates(inplace=True)

# Step 4: Correct Data Types (if needed)
# - "Year" should already be an integer; if not, cast it.
summer_df['Year'] = summer_df['Year'].astype(int)

# Check the processed data
print("Processed Summer Data:\n", summer_df.head())
```

3. Winter Dataset Preprocessing

```python
# Load the Winter dataset
winter_df = pd.read_csv(r"C:\Users\kushi\Olympic\Olympic_Last_attemp\archive\winter.csv")

# Step 1: Standardize Country Codes
# - Remove any trailing spaces from the "Country" column.
winter_df['Country'] = winter_df['Country'].str.strip()

# Step 2: Remove Duplicates
# - Drop duplicates in the dataset.
winter_df.drop_duplicates(inplace=True)

# Step 3: Correct Data Types (if needed)
# - Ensure the "Year" column is an integer type.
winter_df['Year'] = winter_df['Year'].astype(int)

# Check the processed data
print("Processed Winter Data:\n", winter_df.head())
```

```python
import pandas as pd
import re

# Load the CSV file
file_path = r"C:\Users\kushi\Olympic\Olympic_Last_attemp\Fully_Cleaned_Medals.csv"
df = pd.read_csv(file_path, encoding='utf-8')

# Define a function to clean and normalize text in the Medal column
def clean_text(text):
    if isinstance(text, str):
        # Remove non-ASCII characters
        text = re.sub(r'[^\x00-\x7F]+', '', text)
        # Remove any extra spaces or unusual formatting
        text = text.strip()
    return text

# Apply cleaning to the Medal column specifically
df['Medal'] = df['Medal'].apply(clean_text)

# Save to a new file to ensure it's fully cleaned
cleaned_file_path = r"C:\Users\kushi\Olympic\Olympic_Last_attemp\Final_Cleaned_Medals.csv"
df.to_csv(cleaned_file_path, index=False, encoding='utf-8')
```
